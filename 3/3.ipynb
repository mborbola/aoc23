```python
schematic = []

with open('input.txt', 'r') as file:
    for line in file:
        schematic.append(line.strip())
print(schematic[0:5])
```

```python
def is_symbol(char: str):
    return not (char.isdigit() or char == '.')

def get_full_number(s, index) -> int:
    # Ensure index is within the string length
    if index >= len(s) or index < 0:
        return None

    # If the index is on a dot, return None
    if s[index] == '.':
        return None

    # Find the start of the number
    start = index
    while start > 0 and s[start - 1].isdigit():
        start -= 1

    # Find the end of the number
    end = index
    while end < len(s) - 1 and s[end + 1].isdigit():
        end += 1

    # Extract and return the number
    return int(s[start:end + 1])

def get_adjacent_part_numbers(schematic, x, y) -> list:
    numbers = []
    for i in range(x-1, x+2):
        for j in range(y-1, y+2):
            if schematic[i][j].isdigit():
                number = get_full_number(schematic[i], j)
                numbers.append(number)
    return list(set(numbers))
            
def get_part_numbers(schematic):
    part_numbers = []
    for i, line in enumerate(schematic):
        for j, char in enumerate(line):
            if is_symbol(char):
                part_numbers.extend(get_adjacent_part_numbers(schematic, i, j))
    return part_numbers
```

```python
def list_all_numbers(schematic):
    num_list = []
    for i in range(len(schematic)):
        for j in range(len(schematic[i])):
            if schematic[i][j].isdigit():
                num_list.append(get_full_number(schematic[i], j))
    return list(set(num_list))

part_numbers = get_part_numbers(schematic)
num_list = list_all_numbers(schematic)
not_engine_parts = [item for item in num_list if item not in part_numbers]
print(part_numbers)
print(not_engine_parts)
```

```python
def is_gear(char: str):
    return char == '*'

def get_gear_ratios(schematic):
    gear_ratios = []
    for i, line in enumerate(schematic):
        for j, char in enumerate(line):
            if is_gear(char):
                part_numbers = get_adjacent_part_numbers(schematic, i, j)
                if len(part_numbers) == 2:
                    gear_ratio = part_numbers[0] * part_numbers[1]
                    gear_ratios.append(gear_ratio)
    return gear_ratios


test_gear_ratios = get_gear_ratios(test_schematic)
print(test_gear_ratios)
print(sum(test_gear_ratios))
```
